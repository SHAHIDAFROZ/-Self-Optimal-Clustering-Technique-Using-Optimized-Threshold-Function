```python
 #TABLE 4 
    #IMC-2 code
#In this section i have calculated GSI and PI values of IMC2
    
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt



# Specify the path to the image file
image_path = 'C:/Users/shahi/Downloads/p2.jpg'  # Change this to your image file path

# Read the JPEG image
rgb_image = np.array(Image.open(image_path))

# Get the dimensions of the image
rows, cols, _ = rgb_image.shape

# Reshape the RGB image matrix to store the RGB values in x
x = rgb_image.reshape(-1, 3).astype(float)  # Convert to float for precise calculations

# Normalize the data for each dimension of hyperspace
x_min = np.min(x, axis=0)  # Minimum value across all dimensions
x_max = np.max(x, axis=0)  # Maximum value across all dimensions

# Calculate the normalized instances
n = x.shape[0]  # Total number of instances
normalized_instances = (x - np.tile(x_min, (n, 1))) / np.tile((x_max - x_min), (n, 1))
  # Calculate x minimum for every row and store it into H
H = np.min(normalized_instances, axis=1)  # Calculate the minimum along the second dimension (rows)
# Calculate the sum of all values for each row in x and store it into J
J = np.sum(normalized_instances, axis=1)  # Calculate the sum along the second dimension (rows)



delta = 0
M=2
for i in range(len(H)):
    if J[i] != 0:
        delta += (1/(2*n)) * H[i]/J[i] *(M/(M+1))


# Define Q as a unity matrix
Q = np.eye(normalized_instances.shape[1])

# Calculate the distance squared between each pair of points
dist_squared = np.zeros(n)
Pr = np.zeros(n)

for r in range(n):
    for j in range(n):
        diff = normalized_instances[r, :] - normalized_instances[j, :]
        dist_squared[j] = np.dot(np.dot(diff, Q), diff)

    # Compute the potential value Pr for point r
    Pr[r] = np.sum(np.exp(-dist_squared / delta ** 2))  # Assuming the 1st cluster

# Find maximum value of Pr and corresponding point x
max_Pr, max_r = np.max(Pr), np.argmax(Pr)

# Retrieve the corresponding point x from normalized_instances
corresponding_x = normalized_instances[max_r, :]

# Assign above corresponding_x as cluster center
cluster_center = corresponding_x

# Assign data points to the cluster
assigned_cluster_indices = np.zeros(n)  # Initialize assigned cluster indices

for r in range(n):
    # Calculate the Euclidean distance squared between the data point and the cluster center
    dist_squared = np.sum((normalized_instances[r, :] - cluster_center) ** 2)

    # Check if the distance is less than delta
    if dist_squared <= delta:
        assigned_cluster_indices[r] = 1  # Assign the data point to the cluster

# Extract the indices of data points assigned to the cluster
assigned_data_indices = np.where(assigned_cluster_indices == 1)[0]

# Create a logical array indicating which data points are not assigned to the cluster
not_assigned_logical = ~assigned_cluster_indices.astype(bool)

# Extract the data points not assigned to the cluster
y = normalized_instances[not_assigned_logical, :]







# Calculate the total number of instances in cluster 1

m = y.shape[0]

# Calculate x minimum for every row and store it into H1
H1 = np.min(y, axis=1)

# Calculate the sum of all values for each row in y and store it into J1
J1 = np.sum(y, axis=1)

# Initialize beta1

delta1 = 0
for i in range(len(H1)):
    if J1[i] != 0:
        delta1 += (1/(2*m)) * H1[i]/J1[i] * (M/(M+1))


# Define Q as a unity matrix
Q = np.eye(y.shape[1])

# Calculate the distance squared between each pair of points
dist_squared1 = np.zeros(m)
Pr1 = np.zeros(m)

for r in range(m):
    for j in range(m):
        diff1 = y[r, :] - y[j, :]
        dist_squared1[j] = np.dot(np.dot(diff1, Q), diff1)

    # Compute the potential value Pr for point r
    Pr1[r] = np.sum(np.exp(-dist_squared1 / delta1 ** 2))  # Assuming the 1st cluster

# Find maximum value of Pr and corresponding point y
max_Pr1, max_r1 = np.max(Pr1), np.argmax(Pr1)

# Retrieve the corresponding point y from normalized_instances
corresponding_y = y[max_r1, :]

# Assign above corresponding_y as cluster center
cluster_center1 = corresponding_y

# Assign data points to the cluster
assigned_cluster_indices1 = np.zeros(m)  # Initialize assigned cluster indices

for r in range(m):
    # Calculate the Euclidean distance squared between the data point and the cluster center
    dist_squared1 = np.sum((y[r, :] - cluster_center1) ** 2)

    # Check if the distance is less than delta
    if dist_squared1 <= delta1:
        assigned_cluster_indices1[r] = 1  # Assign the data point to the cluster

# Extract the indices of data points assigned to the cluster
assigned_data_indices1 = np.where(assigned_cluster_indices1 == 1)[0]

# Create a logical array indicating which data points are not assigned to the cluster
not_assigned_logical1 = ~assigned_cluster_indices1.astype(bool)

# Extract the data points not assigned to the cluster
z = y[not_assigned_logical1, :]
 # Calculate the distances of each point in z from both cluster centers
distances_to_center = np.sum((z - cluster_center) ** 2, axis=1)
distances_to_center1 = np.sum((z - cluster_center1) ** 2, axis=1)

# Initialize assigned cluster indices for points in z
assigned_cluster_indices = np.zeros(z.shape[0], dtype=int)

# Assign points in z to the cluster with the closest center
for i in range(z.shape[0]):
    if distances_to_center[i] < distances_to_center1[i]:
        assigned_cluster_indices[i] = 1  # Assign to cluster 1
    else:
        assigned_cluster_indices[i] = 2  # Assign to cluster 2

# Extract the indices of data points assigned to each cluster
assigned_data_indices_cluster1 = np.where(assigned_cluster_indices == 1)[0]
assigned_data_indices_cluster2 = np.where(assigned_cluster_indices == 2)[0]

# Data points in cluster 1 (including previously assigned points)
data_points_cluster1 = np.vstack((normalized_instances[assigned_data_indices, :], z[assigned_data_indices_cluster1, :]))

# Data points in cluster 2 (including previously assigned points)
data_points_cluster2 = np.vstack((normalized_instances[assigned_data_indices1, :], z[assigned_data_indices_cluster2, :]))

# Calculate the average distance of each data point to other points in its cluster for cluster 1
a1 = np.zeros(data_points_cluster1.shape[0])
for i in range(data_points_cluster1.shape[0]):
    a1[i] = np.mean(np.sqrt(np.sum((data_points_cluster1 - data_points_cluster1[i, :]) ** 2, axis=1)))

# Calculate the average distance of each data point to points in the nearest neighboring cluster (cluster 2) for cluster 1
b1 = np.zeros(data_points_cluster1.shape[0])
for i in range(data_points_cluster1.shape[0]):
    b1[i] = np.mean(np.sqrt(np.sum((data_points_cluster2 - data_points_cluster1[i, :]) ** 2, axis=1)))

# Calculate the silhouette index for cluster 1
s1 = np.mean((b1 - a1) / np.maximum(a1, b1))

# Calculate the average distance of each data point to other points in its cluster for cluster 2
a2 = np.zeros(data_points_cluster2.shape[0])
for i in range(data_points_cluster2.shape[0]):
    a2[i] = np.mean(np.sqrt(np.sum((data_points_cluster2 - data_points_cluster2[i, :]) ** 2, axis=1)))

# Calculate the average distance of each data point to points in the nearest neighboring cluster (cluster 1) for cluster 2
b2 = np.zeros(data_points_cluster2.shape[0])
for i in range(data_points_cluster2.shape[0]):
    b2[i] = np.mean(np.sqrt(np.sum((data_points_cluster1 - data_points_cluster2[i, :]) ** 2, axis=1)))

# Calculate the silhouette index for cluster 2
s2 = np.mean((b2 - a2) / np.maximum(a2, b2))

# Compute the overall mean silhouette index across all clusters
S = np.mean([s1, s2])


# Calculate the fuzzy cardinality for each cluster
N1 = data_points_cluster1.shape[0]  # Fuzzy cardinality of cluster 1
N2 = data_points_cluster2.shape[0]  # Fuzzy cardinality of cluster 2

# Calculate the centroid of each cluster
centroid1 = np.mean(data_points_cluster1, axis=0)  # Centroid of cluster 1
centroid2 = np.mean(data_points_cluster2, axis=0)  # Centroid of cluster 2

# Calculate the sum of compactness and separation for each cluster
compactness1 = np.sum(np.sqrt(np.sum((data_points_cluster1 - np.tile(centroid1, (N1, 1))) ** 2, axis=1)))  # Compactness of cluster 1
compactness2 = np.sum(np.sqrt(np.sum((data_points_cluster2 - np.tile(centroid2, (N2, 1))) ** 2, axis=1)))  # Compactness of cluster 2

separation1 = np.sum(np.sqrt(np.sum((data_points_cluster1 - np.tile(centroid2, (N1, 1))) ** 2, axis=1)))  # Separation of cluster 1 from cluster 2
separation2 = np.sum(np.sqrt(np.sum((data_points_cluster2 - np.tile(centroid1, (N2, 1))) ** 2, axis=1)))  # Separation of cluster 2 from cluster 1

# Calculate the Partition Index (PI)
PI1 = compactness1 / (N1 * separation1)
PI2 = compactness2 / (N2 * separation2)

PI = PI1 + PI2  # Overall Partition Index

print("IMC2-----GSI = ",S)
print("IMC2-----PI=",PI)

```

```python
 #TABLE 4
#K-means
#In this section i have calculated GSI and PI values of K-means

import numpy as np
from scipy.spatial import distance

# Specify the path to the image file
image_path = 'C:/Users/shahi/Downloads/p2.jpg'  # Change this to your image file path

# Read the JPEG image
rgb_image = plt.imread(image_path)

# Get the dimensions of the image
rows, cols, _ = rgb_image.shape

# Reshape the RGB image matrix to store the RGB values in x
x = rgb_image.reshape(rows * cols, 3).astype(float)  # Convert to float for precise calculations

# Normalize the data for each dimension of hyperspace
x_min = np.min(x, axis=0)  # Minimum value across all dimensions
x_max = np.max(x, axis=0)  # Maximum value across all dimensions

# Calculate the normalized instances
n = x.shape[0]  # Total number of instances
normalized_instances = (x - np.tile(x_min, (n, 1))) / np.tile((x_max - x_min), (n, 1))

# K-means clustering parameters
k = 2  # Number of clusters
max_iterations = 100  # Maximum number of iterations

# Initialize cluster centers
initial_centers = np.array([[0, 0, 0], [0.5, 0.5, 0.5]])  # Initial cluster centers

# Initialize cluster assignments
cluster_assignments = np.zeros(n)

# Initialize cluster centroids
centroids = initial_centers.copy()

# Iterate until convergence or maximum iterations reached
for iter in range(max_iterations):
    # Assign instances to the nearest cluster centroid
    for i in range(n):
        # Calculate distances from instance to centroids
        distances = np.sqrt(np.sum((np.tile(normalized_instances[i, :], (k, 1)) - centroids) ** 2, axis=1))
        # Assign instance to the nearest centroid
        cluster_assignments[i] = np.argmin(distances)
    
    # Update cluster centroids
    for j in range(k):
        # Find instances assigned to cluster j
        cluster_instances = normalized_instances[cluster_assignments == j, :]
        # Update centroid of cluster j
        centroids[j, :] = np.mean(cluster_instances, axis=0)

# Assign each instance to the nearest final cluster centroid
final_cluster_assignments = np.zeros(n)
for i in range(n):
    # Calculate distances from instance to final centroids
    distances = np.sqrt(np.sum((np.tile(normalized_instances[i, :], (k, 1)) - centroids) ** 2, axis=1))
    # Assign instance to the nearest centroid
    final_cluster_assignments[i] = np.argmin(distances)

data_points_cluster1 = normalized_instances[final_cluster_assignments == 0, :]
data_points_cluster2 = normalized_instances[final_cluster_assignments == 1, :]

# Calculate the silhouette index for each cluster
a1 = np.zeros(data_points_cluster1.shape[0])
for i in range(data_points_cluster1.shape[0]):
    a1[i] = np.mean(distance.cdist(data_points_cluster1, [data_points_cluster1[i, :]], 'euclidean'))

b1 = np.zeros(data_points_cluster1.shape[0])
for i in range(data_points_cluster1.shape[0]):
    b1[i] = np.mean(distance.cdist(data_points_cluster2, [data_points_cluster1[i, :]], 'euclidean'))

s1 = np.mean((b1 - a1) / np.maximum(a1, b1))

a2 = np.zeros(data_points_cluster2.shape[0])
for i in range(data_points_cluster2.shape[0]):
    a2[i] = np.mean(distance.cdist(data_points_cluster2, [data_points_cluster2[i, :]], 'euclidean'))

b2 = np.zeros(data_points_cluster2.shape[0])
for i in range(data_points_cluster2.shape[0]):
    b2[i] = np.mean(distance.cdist(data_points_cluster1, [data_points_cluster2[i, :]], 'euclidean'))

s2 = np.mean((b2 - a2) / np.maximum(a2, b2))

S = np.mean([s1, s2])

# Calculate the fuzzy cardinality for each cluster
N1 = data_points_cluster1.shape[0]
N2 = data_points_cluster2.shape[0]

# Calculate the centroid of each cluster
centroid1 = np.mean(data_points_cluster1, axis=0)
centroid2 = np.mean(data_points_cluster2, axis=0)

# Calculate the sum of compactness and separation for each cluster
compactness1 = np.sum(np.sqrt(np.sum((data_points_cluster1 - np.tile(centroid1, (N1, 1))) ** 2, axis=1)))
compactness2 = np.sum(np.sqrt(np.sum((data_points_cluster2 - np.tile(centroid2, (N2, 1))) ** 2, axis=1)))

separation1 = np.sum(np.sqrt(np.sum((data_points_cluster1 - np.tile(centroid2, (N1, 1))) ** 2, axis=1)))
separation2 = np.sum(np.sqrt(np.sum((data_points_cluster2 - np.tile(centroid1, (N2, 1))) ** 2, axis=1)))

# Calculate the Partition Index (PI)
PI1 = compactness1 / (N1 * separation1)
PI2 = compactness2 / (N2 * separation2)

PI = PI1 + PI2
print("------K-means---------")

print(f'Overall Silhouette Index (GSI): {S}')
print(f'Partition Index (PI): {PI}')
```
